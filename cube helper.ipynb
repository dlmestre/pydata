```python
import pandas as pd

path = r"/home/ubuntu/Files/SUBNATIONAL/subnational.csv"
DF = pd.read_csv(path)
len(DF)
```

```python
DF = DF[DF["Indicator Name"]=="Poverty headcount ratio at national poverty line (% of population)"]
DF_Countries = DF["Country Name"]
len(DF_Countries)
```

```python
Countries = DF["Country Name"].tolist()
Countries = Countries*len(range(1996,2014))
Years = []

def replacer(string):
    #return string.replace("\xe1","á").replace("\xed","í").replace("\xfd","ý").replace("\xe4","ä").replace("\xe9","é"). \
    #replace("\xe8","è")
    return string.replace("\xe1","a").replace("\xed","i").replace("\xfd","y").replace("\xe4","a").replace("\xe9","e"). \
    replace("\xe8","e").replace("\xe2","a").replace("\xf3","o")

for year in range(1996,2014):
    Year = [year]*858
    for k in Year:
        Years.append(k)

        
Countries_names = [Country.split(",")[0] for Country in Countries]
Regions = [replacer("".join(Country.split(",")[1:]).strip()) for Country in Countries]

poverty = []
for year in range(1996,2014):
    for j in DF[str(year)]:
        poverty.append(j)

new_df = pd.DataFrame({"Countries":Countries_names,
                       "Regions":Regions,
                       "Years":Years,
                       "Poverty Ratio":poverty})
new_df.dropna()
new_df.head()

def set_columns_order(df,order):
    cols = order[:]
    for col in df.columns:
        if col not in cols:
            cols.append(col)
    return df[cols]

df = set_columns_order(new_df,["Countries","Regions","Years","Poverty Ratio"])
df.head()
```
