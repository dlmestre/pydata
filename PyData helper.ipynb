```python
from __future__ import print_function
from cubes import Workspace, Cell, PointCut

path_file = r"/home/ubuntu/Files/examples/models.json"

workspace = Workspace()
workspace.register_default_store("sql", url="sqlite:///data.sqlite")
workspace.import_model(path_file)

browser = workspace.browser("example_file")
```

```python
r = browser.aggregate(drilldown=["item"])
for row in r.table_rows("item"):
    print(row.label)
```

```python
cut = PointCut("item", ["e"])
cell = Cell(browser.cube, cuts = [cut])

result = browser.aggregate(cell, drilldown=["Year"])
print(dir(result))
for record in result:
    print(record)
```

```python
import sqlite3
con = sqlite3.connect('data.sqlite')

with con:
    
    cur = con.cursor() 
    cur.execute("SELECT name FROM sqlite_master WHERE type='table';")    
    tables = cur.fetchall() 
    print(tables)

    for table_name in tables:
        print(table_name)
```

```python
from sqlalchemy import create_engine
from cubes.tutorial.sql import create_table_from_csv

engine = create_engine('sqlite:///pydata.sqlite')

path_life_expectancy = r"/home/ubuntu/Files/WDI/WDI_life_expectancy.csv"
path_gdp_per_capita = r"/home/ubuntu/Files/GAPDATA/gapdata.csv"

create_table_from_csv(engine,
                      path_life_expectancy,
                      table_name="wdi_life_expectancy",
                      fields=[("Countries", "string"),
                              ("Life Expectancy", "string"),
                              ("Years", "integer")],
                      create_id=True)

create_table_from_csv(engine,
                      path_gdp_per_capita,
                      table_name="gap_gdp_per_Capita",
                      fields=[("Area", "string"),
                              ("GDP per capita", "integer"),
                              ("Year", "integer")],
                      create_id=True)
```
